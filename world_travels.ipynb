```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the world map data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

countries=world['name'].sort_values()
for name in countries:
  print(name)
```

```python
countries_visited= ['India', 'Canada', 'Iran', 'Iraq', 'Kuwait', 'Saudi Arabia', 'Jordan', 'Syria', 'Lebanon', 'Libya', 'Morocco', 'Egypt', 'Sudan', 'Uganda',  'Kenya', 'Tanzania', 'Zambia', 'Malawi', 'Botswana', 'South Africa', 'Nigeria', 'Liberia',  'Ghana', 'Russia', 'Italy', 'Greece', 'Malta', 'Turkey', 'Cyprus', 'United Kingdom', 'France',  'Switzerland', 'Germany', 'Denmark', 'United States of America', 'Mexico', 'Oman', 'Yemen', 'Turkey', 'Benin', 'Togo', 'Mozambique', 'Zimbabwe', 'Germany', 'Netherlands', 'Belgium', 'Cyprus']
```

```python
# Plot the world map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.plot(ax=ax, color='lightblue', edgecolor='black')

for country in countries_visited:
  try:
    world[world.name == country].plot(edgecolor=u'gray', color='purple', ax=ax)
  except:
    print(country)

ax.set_title('Countries S Visited', fontsize=15)
plt.show()
```
